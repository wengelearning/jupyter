```python
import matplotlib.pyplot as plt#画两个点间的线
import numpy as np
a1=np.array([1,5])
b2=np.array([3,4])
c=np.array([a1,b2])
x=(c.T)[0]
y=(c.T)[1]
x1,y1=a1
x2,y2=b2
plt.scatter(x1,y1,c='r')
plt.scatter(x2,y2,c='b')
plt.plot(x,y,c='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(-2,2,100)
b=a**2
plt.xlabel('I am X')#设置x轴标签
plt.ylabel('I am Y')
plt.yticks([0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0],['lv1','lv2','lv3','lv4','lv5','lv6','lv7','lv8','lv9'])#设置y轴刻度显示
plt.plot(a,b,color='r',linestyle='--',linewidth=2)#画线，并设置线条颜色，格式，粗细
plt.show()
```
